### ClassifierBasedGermanTagger

```python
import nltk
import re
from datetime import datetime
from nltk.corpus.reader.conll import *
```

```python
#open a file with german text for processing
with open('deu-text.txt','r', encoding = "utf-8") as f:
    text = f.read()
```

```python
print(text)
```

```python
# maybe we will need tokenization for our pos-taggers, just let's have this function
def tokenize(one_new):
    tokens = one_new.replace(' - ', ' ').replace('. ', ' ')
    tokens = re.sub(r'\d+', '', tokens)
    umlya = re.compile(r'[\w]+')
    all_tokens = umlya.findall(tokens)
    return all_tokens
```

```python
y = tokenize(text)
y
```

```python
# for reading our corpus 
corp = nltk.corpus.ConllCorpusReader('.', 'tiger_release_aug07.corrected.16012013.conll09',
                                     ['ignore', 'words', 'ignore', 'ignore', 'pos'],
                                     encoding='utf-8')
```

```python
import random,math
#we need to separate data for training and testing
tagged_sents = corp.tagged_sents()
i = math.floor(len(tagged_sents)*0.2)
testing_sentences = tagged_sents[0:i]
training_sentences = tagged_sents[i:]
```

```python
#now we train it (we can save model with the help of pickle module, but I didn't)
from ClassifierBasedGermanTagger import ClassifierBasedGermanTagger
tagger = ClassifierBasedGermanTagger(train=training_sentences) 
```

#### Evaluation (accuracy & time)

```python
accuracy = tagger.evaluate(testing_sentences)
```

```python
accuracy
```

```python
%timeit result_1 = tagger.tag(y)
```

```python
tagger.tag(y)
```

```python
start_time = datetime.now()
tagger.tag(y)
end_time = datetime.now()
print('Duration: {}'.format(end_time - start_time))
```
